```python
import json, os, pprint, time, re
from pprint import pprint as pp
from urllib import parse
import requests as req
from bs4 import BeautifulSoup as bs
import random
from selenium import webdriver
```

```python
# 隨機取得user agent
from fake_useragent import UserAgent
ua = UserAgent(cache=True)

print(ua.random)
```

```python
# https://www.haodoo.net/?M=hd&P=history-1
# https://www.haodoo.net/?M=hd&P=history-2
# https://www.haodoo.net/?M=hd&P=history-3
# https://www.haodoo.net/?M=hd&P=100-{1-4}
my_headers = {
    'user-agent': ua.random
}
bookList = []

folderPath = 'haodoo'
if not os.path.exists(folderPath):
    os.makedirs(folderPath)

url = "https://www.haodoo.net/?M=hd&P=100-5"
response = req.get(url, headers = my_headers)
soup = bs(response.text, 'lxml')
#print(soup)
a_elements = soup.select('div[class="a03"] a')
print(a_elements[0])

for a in a_elements:
    eachBookLink = {
        "bookTitle" : re.sub("【|】|《|》|\n", "",a.get_text()),
        "link" : "https://www.haodoo.net" + parse.unquote(a.get('href'))
        
    }
    bookList.append(eachBookLink)
pp(bookList)
    
```

```python
# get links on each page
# Dowload Link : https://www.haodoo.net/?M=d&P=C1296.epub

'''
find_previous(name=None, attrs={}, text=None, **kwargs)
where
name -A filter on tag name.
attrs - A dictionary of filters on attribute values.
text - A filter for a NavigableString with specific text.
kwargs - A dictionary of filters on attribute values.

'''

for book in bookList:
    
    responseEachBook = req.get(book['link'], headers = my_headers)
    soupEachBook = bs(responseEachBook.text, 'lxml')

# print(type(str(soupEachBook)))
#pp(soupEachBook)

    a_list = soupEachBook.find_all(name='input',value="線上閱讀")
    print(a_list)

    for a in a_list:
        author_title = a.find_previous(name="font").get_text()
        authpath = f"{folderPath}/{author_title}"
        book_title = a.find_next(name='input', value="下載 epub 檔")
        print(authpath)
        dlfilename=book_title['onclick'].replace("DownloadEpub('","").replace("')","")+".epub"
        #print(author_title)
        print(dlfilename)
        if not os.path.exists(authpath):
            os.makedirs(authpath)
        #print(author_title)
        #print(book_title)
    
        dllink = f"https://www.haodoo.net/?M=d&P={dlfilename}"
        dlfile = req.get(dllink, allow_redirects=True)
        savefile = open(authpath+"/"+dlfilename, 'wb')
        savefile.write(dlfile.content)
        savefile.close()
    

    
```
